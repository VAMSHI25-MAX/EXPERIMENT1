```python
class Animal:
    def speak(self):
        return " different sound"
class dog(Animal):
    def speak(self):
        return "bark"
class cat(Animal):
    def speak(self):
        return "meow"
detail=[dog(), cat(),Animal()]
for data in detail:
    print(data.speak())
```

```python
class vehicle:
    def cost(self):
        return "cost of vehicle"
class bike(vehicle):
    def cost(self):
        return "bike: 1lakh"
class car(vehicle):
    def cost(self):
        return "car: 10lakhs"
class bus(vehicle):
    def cost(self):
        return "bus: 50lakhs"
class aeroplane(vehicle):
    def cost(self):
        return "aeroplane: 10crores"
data=[vehicle(),bike(),car(),bus(),aeroplane()]
for i in data:
    print(i.cost())
```

```python
  class Employee:
      def work(self, hours=None):
          if hours is None:
              print("Employee working default 8 hours")
          else:
              print("Employee working {hours} hours today")
  class Nurse(Employee):
      # Overload work method
      def work(self, hours=None, shift=None):
          if hours is None:
              print("Nurse working default 8 hours")
          elif shift is None:
              print(f"Nurse working {hours} hours today")
          else:
              print(f"Nurse working {hours} hours on {shift} shift") 
```
